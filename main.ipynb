Data Cleaning

```python
# clean Ticket_price_cr

df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].astype(str)
```

```python
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].str.replace('Cr','',regex=False).str.strip()
```

```python
df['Ticket_Price_Cr'] = pd.to_numeric(df['Ticket_Price_Cr'], errors='coerce')
```

```python
df['Ticket_Price_Cr'].dtype
```

```python
## Handle missing value 

df.isnull().sum()
```

```python
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].fillna(df['Ticket_Price_Cr'].median())
```

```python
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].fillna(df['Ticket_Price_Cr'].median())

df['Unit_Size_Sqft'] = df['Unit_Size_Sqft'].fillna(df['Unit_Size_Sqft'].median())

df['Amenity_Score'] = df['Amenity_Score'].fillna(df['Amenity_Score'].median())

df['Buyer_Comments'] = df['Buyer_Comments'].fillna('No comments')
```

```python
df.isnull().sum()
```

 Feature Engineering

```python
## Booking flag 

df['Booking_Flag'] = df['Transaction_Type'].apply(
    lambda x: 1 if x == 'Primary' else 0
)
```

```python
df['Booking_Flag'].mean()
```

```python
df['Booking_Flag'].value_counts()
```

```python
## price per sqft

df['Price_per_Sqft'] = (df['Ticket_Price_Cr'] * 10000000) / df['Unit_Size_Sqft']
```

```python
df['Price_per_Sqft']
```

```python
df['Purchase_Quarter'].unique()
```

```python
## Quarter Number 

# convert to datetime 
df['Purchase_Quarter'] = pd.to_datetime(df['Purchase_Quarter'])

# extract quarter 
df['Quarter_Number'] = df['Purchase_Quarter'].dt.quarter
    
```

```python
df[['Purchase_Quarter', 'Quarter_Number']].head()
```

```python
## price category 

df['Price_Category'] = pd.cut(
    df['Ticket_Price_Cr'],
    bins=[0, 2, 5, 10, 100],
    labels=['low', 'Mid', 'High', 'Ultra Luxury']
)
```

```python
## Premium Property

df['Premium_Property'] = df['Amenity_Score'].apply(
    lambda x: 1 if x > 8 else 0
)
```
